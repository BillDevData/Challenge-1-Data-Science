### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda4.shape
```

```python
tienda4.info()
```

#1. Análisis de facturación



```python
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

print(f'Ingreso de la tienda 1: {ingreso_tienda_1}')
print(f'Ingreso de la tienda 2: {ingreso_tienda_2}')
print(f'Ingreso de la tienda 3: {ingreso_tienda_3}')
print(f'Ingreso de la tienda 4: {ingreso_tienda_4}')
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Producto"].count().reset_index().rename(columns={"Producto": "Cantidad Vendida"}).sort_values(by="Cantidad Vendida", ascending=False)
ventas_categoria_tienda1
```

```python
ventas_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Producto"].count().reset_index().rename(columns={"Producto": "Cantidad Vendida"}).sort_values(by="Cantidad Vendida", ascending=False)
ventas_categoria_tienda2
```

```python
ventas_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Producto"].count().reset_index().rename(columns={"Producto": "Cantidad Vendida"}).sort_values(by="Cantidad Vendida", ascending=False)
ventas_categoria_tienda3
```

```python
ventas_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Producto"].count().reset_index().rename(columns={"Producto": "Cantidad Vendida"}).sort_values(by="Cantidad Vendida", ascending=False)
ventas_categoria_tienda4
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda1 = tienda["Calificación"].mean()
print("Promedio Tienda 1:", promedio_tienda1)

promedio_tienda2 = tienda2["Calificación"].mean()
print("Promedio Tienda 2:", promedio_tienda2)

promedio_tienda3 = tienda3["Calificación"].mean()
print("Promedio Tienda 3:", promedio_tienda3)

promedio_tienda4 = tienda4["Calificación"].mean()
print("Promedio Tienda 4:", promedio_tienda4)
```

# 4. Productos más y menos vendidos

```python
ventas_productos_tienda1 = tienda.groupby("Producto")["Producto"].count().reset_index(name="Cantidad Vendida").sort_values(by="Cantidad Vendida", ascending=False)
ventas_productos_tienda2 = tienda2.groupby("Producto")["Producto"].count().reset_index(name="Cantidad Vendida").sort_values(by="Cantidad Vendida", ascending=False)
ventas_productos_tienda3 = tienda3.groupby("Producto")["Producto"].count().reset_index(name="Cantidad Vendida").sort_values(by="Cantidad Vendida", ascending=False)
ventas_productos_tienda4 = tienda4.groupby("Producto")["Producto"].count().reset_index(name="Cantidad Vendida").sort_values(by="Cantidad Vendida", ascending=False)
```

```python
ventas_tiendas = [
    ("Tienda 1", ventas_productos_tienda1),
    ("Tienda 2", ventas_productos_tienda2),
    ("Tienda 3", ventas_productos_tienda3),
    ("Tienda 4", ventas_productos_tienda4),
]

for nombre, data in ventas_tiendas:
    max_ventas = data["Cantidad Vendida"].max()
    min_ventas = data["Cantidad Vendida"].min()

    mas_vendidos = data[data["Cantidad Vendida"] == max_ventas]
    menos_vendidos = data[data["Cantidad Vendida"] == min_ventas]

    print('********************************************************')
    print(f"{nombre}")
    print("\nMás vendidos:")
    print(mas_vendidos)
    print("\nMenos vendidos:")
    print(menos_vendidos)
    print('\n********************************************************')
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda = tienda["Costo de envío"].mean().round(2)
envio_promedio_tienda2 = tienda2["Costo de envío"].mean().round(2)
envio_promedio_tienda3 = tienda3["Costo de envío"].mean().round(2)
envio_promedio_tienda4 = tienda4["Costo de envío"].mean().round(2)

print("Tienda 1:", envio_promedio_tienda)
print("Tienda 2:", envio_promedio_tienda2)
print("Tienda 3:", envio_promedio_tienda3)
print("Tienda 4:", envio_promedio_tienda4)
```

# GRÁFICOS

```python
import matplotlib.pyplot as plt
```

1. Ingresos por tienda

```python
ingresos_tiendas = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingreso Total": [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
}

df_ingresos = pd.DataFrame(ingresos_tiendas)
df_ingresos
```

```python
plt.figure(figsize=(8,5))
plt.bar(df_ingresos["Tienda"], df_ingresos["Ingreso Total"], color="blue")

plt.title("Ingresos Totales por Tienda", fontsize=14, fontweight="bold")
plt.xlabel("Tienda")
plt.ylabel("Ingreso Total")

plt.show()
```

2. Calificación por tienda

```python
calificacion_tiendas = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Calificación Promedio": [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]
}

df_calificacion = pd.DataFrame(calificacion_tiendas)
df_calificacion
```

```python
plt.figure(figsize=(8,5))
bars = plt.barh(df_calificacion["Tienda"], df_calificacion["Calificación Promedio"], color="green")

plt.axvline(x=4.0, color="red", linestyle="--", linewidth=1.5, label="Referencia 4.0")

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.05, bar.get_y() + bar.get_height()/2, f"{width:.2f}", va="center", fontsize=10, fontweight="bold")

plt.title("Calificación Promedio de clientes por Tienda (1 a 5 estrellas)", fontsize=14, fontweight="bold")
plt.xlabel("Calificación Promedio")
plt.xlim(0, 5)

plt.show()
```

3. Costo de envio promedio por tienda

```python
envio_promedio_tiendas = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de envío promedio": [envio_promedio_tienda, envio_promedio_tienda2, envio_promedio_tienda3, envio_promedio_tienda4]
}

df_envio = pd.DataFrame(envio_promedio_tiendas)
df_envio
```

```python
plt.figure(figsize=(8,5))
bars = plt.bar(df_envio["Tienda"], df_envio["Costo de envío promedio"], color="skyblue")

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval,2),
             ha='center', va='bottom', fontsize=10, fontweight="bold")

plt.title("Costo de envío promedio por tienda", fontsize=14, fontweight="bold")
plt.xlabel("Tienda", fontsize=12)
plt.ylabel("Costo de envío promedio", fontsize=12)

plt.show()
```
